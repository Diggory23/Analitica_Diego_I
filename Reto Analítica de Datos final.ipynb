```python
import numpy as np
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
sb.pairplot(new_df.dropna(), hue='user_verified',height=4,vars=["user_followers","user_favourites"],kind='scatter')
```

```python
X = np.array(new_df[["user_favourites","user_followers"]])
y = np.array(new_df['user_verified'])
X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red']
asignar=[]
for row in y:
    asignar.append(colores[row])
ax.scatter(X[:, 0], X[:, 1], c=asignar,s=60)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
labels = kmeans.predict(X)
C = kmeans.cluster_centers_
colores=['red','blue']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1],  c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1],  marker='*', c=colores, s=1000)
```

```python
copy =  pd.DataFrame()
copy['user_verified']=new_df['user_verified'].values
copy['user_favourites']=new_df['user_favourites'].values
copy['label'] = labels;
cantidadGrupo =  pd.DataFrame()
cantidadGrupo['color']=colores
cantidadGrupo['cantidad']=copy.groupby('label').size()
cantidadGrupo
```

```python
group_referrer_index = copy['label'] ==0
group_referrals = copy[group_referrer_index]

diversidadGrupo =  pd.DataFrame()
diversidadGrupo['user_verified']=[0,1]
diversidadGrupo['cantidad']=group_referrals.groupby('user_verified').size()
diversidadGrupo
```
